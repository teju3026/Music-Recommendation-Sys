```python
new_df=df[['Name','Artists']]
new_df
```

```python

new_df['Artists']=new_df['Artists'].apply(lambda x:x.lower())
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2000)
```

```python
vectors=cv.fit_transform(new_df['Artists']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
new_df.rename(columns={'Name':'Artists'},inplace=True)
```

```python
def recommend(music):
    music_index=new_df[new_df['Artists']==music].index[0]
    distances=similarity[music_index]
    music_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in music_list:
        print(new_df.iloc[i[0]].Artists)

```

```python
recommend('Gabbar Bhi Nachega - Remix	')
```
